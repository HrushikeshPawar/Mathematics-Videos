```python
from manim import *
#from manimlib import *
from random import randint
from random import uniform
```

```python
%%manim VectorArrow

class VectorArrow(Scene):
    def construct(self):

        def midpoint(p1, p2):

          mid = []
          for i in range(len(p1)):
            mid.append((p1[i]+p2[i])*0.5)
          return mid
        
        #a, b, c, d, e = [-2,-2,0], [2, -2, 0], [3,2,0], [0,4,0], [-3,2,0]
        #a, b, c, d = [5,-4,0], [-5,-2.5,0], [-5, 3, 0], [5, 4, 0]
        a, b, c = [5,-4,0], [-5,-4,0], [0, 4, 0]
        poly = Polygon(a,b,c, stroke_width=2)
        self.add(poly)

        coords = [[4,0,0]]
        

        for i in range(10000):

          dice = randint(1, 4)

          if dice == 1:
            coords.append(midpoint(coords[-1], a))

          if dice == 2:
            coords.append(midpoint(coords[-1], b))

          if dice == 3:
            coords.append(midpoint(coords[-1], c))
          """
          if dice == 4:
            coords.append(midpoint(coords[-1], d))
          """

        group = VGroup()
        for point in coords:
          group.add(Dot(point).scale(0.2))
        
        self.add(group)
```

```python
%%manim ThreeD

class ThreeD(ThreeDScene):
  def construct(self):
    
    tracepoint = [0,0,0]
    x_min = -5
    x_max = +5
    y_min = -4
    y_max = 4
    z_min = -4
    z_max = 4
    n = 5
    m = 0.48
    N = 50000

    def get_vertices(n):

      vertices = [[4, 4, -1], [-4, 4, -1], [-4, -4, -1],  [4, -4, -1], [0, 0, 3]]
      for point in vertices:
        self.add(Dot(point, color=PURPLE))

      return vertices
    
    def draw_lines(vertices):

      #vertices = sorted(vertices)

      lines = VGroup()
      for i in range(3):
        lines.add(Line(vertices[i], vertices[i+1], stroke_width=0.5, color=PURPLE))
      lines.add(Line(vertices[0], vertices[3], stroke_width=0.5, color=PURPLE))

      
      for i in range(4, len(vertices)):
        for j in range(4):
          lines.add(Line(vertices[i], vertices[j], stroke_width=0.5, color=PURPLE))

      self.add(lines)
    
    def multiplier(p1, p2, m):

        point = []
        for i in range(len(p1)):
          point.append((p1[i]+p2[i])*m)
        return point
    
    def Chaos(Vertices, tracepoint, N, n, m):

      points = [tracepoint]
      color_pallet = [BLUE, GREEN, YELLOW, ORANGE, RED]
      group = VGroup()
      for i in range(N):

        dice = randint(1, n)

        points.append(multiplier(points[-1], Vertices[dice-1], m))
        group.add(Dot(points[-1], color=color_pallet[dice-1]).scale(0.2))

        if i % 1000 == 0:
          print(f"Points : {i}")        
      
      print(f"Completed Point Calculation")
      return group


    print(f"Setting Up Camera")
    self.set_camera_orientation(phi=75 * DEGREES, theta=30 * DEGREES)
    print(f"Camera Setup Completed")
    Vertices = get_vertices(n)
    draw_lines(Vertices)

    
    self.add(Chaos(Vertices, tracepoint, N, n, m))
    print(f"Added Objects to the Scene")
    print(f"Moving the Camera Now.")
    self.begin_ambient_camera_rotation(rate=0.5)
    self.wait(10)
    self.stop_3dillusion_camera_rotation()
    
```
